```python
import numpy as np
import matplotlib.pyplot as plt
from neuralSLModule import nn_model, predict

%matplotlib inline
np.random.seed(1)
```

```python
# Dataset definition
num = 1000

# Training set
X1 = np.random.randn(1, num) + 3
Y1 = np.random.randn(1, num) + 3

X2 = np.random.randn(1, num) - 3
Y2 = np.random.randn(1, num) - 3

X3 = np.random.randn(1, num) - 0
Y3 = np.random.randn(1, num) - 0

x = np.concatenate((X1, X2, X3), axis=1)
y = np.concatenate((Y1, Y2, Y3), axis=1)

dataset = np.concatenate((x, y), axis=0)
label1 = np.zeros((1, num))
label2 = np.ones((1, num))
label3 = np.ones((1, num))
label = np.concatenate((label1, label2, label3), axis=1)

# Test set
X_test = np.random.randn(1, num)
Y_test = np.random.randn(1, num)
testset = np.concatenate((X_test, Y_test), axis=0)
```

```python
plt.scatter(dataset[0,:], dataset[1,:], c=label.ravel());
```

```python
# Training on dataset
parameters = nn_model(dataset, label, 20, num_iterations = 10000, print_cost=True)
```

```python
# Decision boundary
n = 100000
X_grid = np.random.rand(2, n)*10 - 5
decision_boundary = predict(parameters, X_grid)
plt.scatter(X_grid[0,:], X_grid[1,:], c=decision_boundary.ravel(), s=1);
```

```python
# Test set defintion
X_test = np.random.randn(1, num)
Y_test = np.random.randn(1, num)
testset = np.concatenate((X_test, Y_test), axis=0)
```

```python
# Test set labeling
label_test = predict(parameters, testset)
plt.scatter(testset[0,:], testset[1,:], c=label_test.ravel());
```
